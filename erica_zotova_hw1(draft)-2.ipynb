#### Создайте колонку 'Top_University' со значением True для университетов, входящих в топ-50. Обратите внимание на значения типа "231–240".

#### Создайте колонку Academic_Performance, в которую внесите значение:
#### - "High", если GPA выше 3.5 баллов,
#### - "Medium", если GPA от 3 до 3.49 баллов,
#### - "Low", если GPA ниже 3 баллов.

#### Посмотрите на уникальные значения в колонке Degree и приведите их к единообразию (Bachelor of Science -> Bachelor).

```python
df_succ_edu_2['GPA (or Equivalent)'] = pd.to_numeric(df_succ_edu_2['GPA (or Equivalent)'])
def academic_performance(gpa):
    if isinstance(gpa, float):
        if gpa > 3.5:
            perf = 'High'
        elif gpa >= 3:
            perf = 'Medium'
        elif gpa < 3:
            perf = 'Low'
        return perf
    return None

df_succ_edu_2['Academic_Performance'] = df_succ_edu_2['GPA (or Equivalent)'].apply(academic_performance)
df_succ_edu_2.head(15)
```

```python
df_succ_edu_2['Degree'].unique()
```

```python
#мне не нравится адхоковость этого способа, но на большее я не способна:(

def universal_degree(degree):
    if isinstance(degree, str):
        if 'MD' in degree:
            degree_modified = 'Master'
        elif 'MBA' in degree:
            degree_modified = 'MBA' #ее обычно отдельно выносят, как отдельный вид степени
        elif 'Master' in degree:
            if 'Business' in degree:
                degree_modified = 'MBA'
            else:
                degree_modified = 'Master'
        elif 'Bachelor' in degree:
            degree_modified = 'Bachelor'
        elif 'PhD' in degree or 'Doctor' in degree or 'Ph.D' in degree:
            degree_modified = 'PhD' #обычно пишут так; honorary doctorate я не поняла что такое, пусть оно туда попадает
        else:
            degree_modified = degree
        return degree_modified
    return None

#Undergraduate Studies идейно должно быть в разделе бакалавриата, но чето я засомневалась, пусть будет отдельно

df_succ_edu_2['Degree'] = df_succ_edu_2['Degree'].apply(universal_degree)
```
